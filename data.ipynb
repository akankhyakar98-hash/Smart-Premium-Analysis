```python
import pandas as pd
from scipy.stats import ttest_ind

# 1. Show how many smokers vs. non-smokers you have
print("Counts by Smoking Status:")
print(df["Smoking Status"].value_counts())
print()

# 2. Compute groupâ€level summary statistics on Premium Amount
grouped = df.groupby("Smoking Status")["Premium Amount"] \
            .agg(count="count",
                 mean="mean",
                 median="median",
                 std="std",
                 q1=lambda x: x.quantile(0.25),
                 q3=lambda x: x.quantile(0.75))
print("Premium stats by Smoking Status:")
print(grouped)
print()

# 3. (Optional) Run a simple t-test to see if the means differ significantly
smokers     = df.loc[df["Smoking Status"] == "Yes", "Premium Amount"].dropna()
nonsmokers  = df.loc[df["Smoking Status"] == "No",  "Premium Amount"].dropna()
t_stat, p_val = ttest_ind(smokers, nonsmokers, equal_var=False)

print(f"T-statistic: {t_stat:.2f}, p-value: {p_val:.3f}")
if p_val < 0.05:
    print("â†’ Means are significantly different (p < 0.05).")
else:
    print("â†’ No significant difference in means (p â‰¥ 0.05).")
```

```python
#ANOVA TEST- to determine whether premium is significantly different across groups such as smokers & non-smokers etc
from scipy.stats import f_oneway

# List of categorical features to test
categorical_features = ['Gender', 'Smoking Status', 'Policy Type', 'Exercise Frequency']

print(" ANOVA Test Results:")
for feature in categorical_features:
    groups = [group['Premium Amount'].dropna().values for _, group in df.groupby(feature)]
    f_stat, p_val = f_oneway(*groups)
    print(f"{feature}: p-value = {p_val:.4f} â†’ {'Significant' if p_val < 0.05 else 'Not Significant'}")
```

```python
#OUTLIERS TREATMENT
import seaborn as sns
import matplotlib.pyplot as plt

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
# Loop through numerical columns and create boxplots

for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Check: {col}")
    plt.show()
```

```python
def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series.clip(lower, upper)


for col in ['Premium Amount', 'Annual Income', 'Health Score', 'Vehicle Age', 'Insurance Duration']:
    df[col] = cap_outliers(df[col])
```
